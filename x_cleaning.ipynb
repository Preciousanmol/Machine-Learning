```python
xx = pd.read_csv('d.csv')
xx.head()
```

```python
# split date into 3 columns day, month and year
xx['year']=xx['date'].str.split('-').str.get(0)
xx['year'].head()
```

```python
xx['month']=xx['date'].str.split('-').str.get(1)
xx['month'].head()
```

```python
xx['day']=xx['date'].str.split('-').str.get(2)
xx['day'].head()
```

```python
xx.head()
```

```python
# want data in day0month-year format. Now join day, month and year column

xx['updated_date']=xx['month'] + '-'+ xx['day'] + '-' + xx['year']
xx['updated_date'].head()

# now we have date in same format of y (air_precptation)
```

```python
xx.head()
```

```python
xx = xx.drop(['year','month','day','Unnamed: 0','Unnamed: 0.1','date'],axis=1)

xx.head()
```

```python
xx['hour'] = xx['time'].str.split(':').str.get(0)
xx['hour'].head()

xx['min'] = xx['time'].str.split(':').str.get(1)
xx['min'].head()


xx['time_updated'] = xx['hour'] + ':' + xx['min']
xx.to_csv('refined_x.csv')
xx.head()
```

```python
# join date and time now
xx['updated_timestamp'] = xx['updated_date'] + ' ' + xx['time_updated']

xx['updated_timestamp'].head()

xx.to_csv('refined_x.csv')
xx
```
