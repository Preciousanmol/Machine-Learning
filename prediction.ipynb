```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('refined_x_y.csv')

df.head()
```

# using all features in x (air temp, air pressure, windspeed, specific humidity, timestamp)

```python
x = df.iloc[:,1:5]
x.head()
df

```

```python
y = df.iloc[:,6]
y.head()
y=y.reshape(-1,1)
```

```python
#convert updated_timestamp to float
#from sklearn.preprocessing import LabelEncoder
#X_labelencoder = LabelEncoder()
#x.iloc[:, 4] = X_labelencoder.fit_transform(x.iloc[:, 4])
```

```python
from sklearn.preprocessing import Imputer
# First create an Imputer , Stratergy means what we want to write in place of missed value
missingValueImputer = Imputer (missing_values = 'NaN', strategy = 'mean', 
                               axis = 0)  #if missing values are represented by 9999 then write same here
# Set which columns imputer should perform
missingValueImputer = missingValueImputer.fit (x)
# update values of X with new values
x = missingValueImputer.transform(x)
x

missingValueImputer = Imputer (missing_values = 'NaN', strategy = 'mean', 
                               axis = 0)  #if missing values are represented by 9999 then write same here
# Set which columns imputer should perform
missingValueImputery = missingValueImputer.fit (y)
# update values of X with new values
y = missingValueImputery.transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2)
```

```python
print("Shape of X_train",X_train.shape)
print("Shape of X_test",X_test.shape)
print("Shape of Y_train",Y_train.shape)
print("Shape of Y_test",Y_test.shape)
```

```python
#X_train.info()
```

# third algorithm

```python
from sklearn.linear_model import LinearRegression
regressoragent = LinearRegression()
regressoragent.fit (X_train, Y_train )            
                   
#==============================================================================
# Now check what our model learned by predicting the X_test values
#==============================================================================

predictValues = regressoragent.predict(X_test)

predictValues
```

```python
#actual and predicted y comparison

plt.scatter(X_test[:,0],Y_test, color = 'red')
plt.plot(X_test[:,0],predictValues, color = 'blue')
plt.title('Actual vs Predicted Y')
plt.xlabel('Actual y test')
plt.ylabel('Predicted y test')
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from math import sqrt
print("RMSE is:-")
print(np.sqrt(metrics.mean_squared_error(Y_test, predictValues)))

```

# Model on 2 features - air temperature and sspecific humidity

```python
x_new = df.iloc[:,[3,2]]
x_new.head()
```

```python
from sklearn.preprocessing import Imputer
# First create an Imputer , Stratergy means what we want to write in place of missed value
missingValueImputer = Imputer (missing_values = 'NaN', strategy = 'mean', 
                               axis = 0)  #if missing values are represented by 9999 then write same here
# Set which columns imputer should perform
missingValueImputer = missingValueImputer.fit (x_new)
# update values of X with new values
x_new = missingValueImputer.transform(x_new)


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_new, y, test_size = 0.2)
```

```python
print("Shape of X_train",X_train.shape)
print("Shape of X_test",X_test.shape)
print("Shape of Y_train",Y_train.shape)
print("Shape of Y_test",Y_test.shape)
```

# Training using MLPRegressor

```python
from sklearn.neural_network import MLPRegressor
clf = MLPRegressor(solver='sgd', alpha= 1/1+1e-3, hidden_layer_sizes=(50, 50))
clf.fit(X_train, Y_train)
```

# predicted values for x_test

```python
pred = clf.predict(X_test)
pred

```

```python
np.savetxt('predictions.csv', pred, delimiter=',',fmt='%5s') 
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, pred)
mse
```
