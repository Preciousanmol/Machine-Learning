```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
air_precipitation = pd.read_csv('PRECIPITATIONnn.csv')

air_precipitation.head()
```

```python
air_precipitation['time']=air_precipitation['timestamp'].str.split(' ').str.get(1)
air_precipitation['time'].head()

```

```python
air_precipitation['date']=air_precipitation['timestamp'].str.split(' ').str.get(0)
air_precipitation['date'].head()

air_precipitation['date']=air_precipitation['date'].str.replace('/','-')
air_precipitation['date'].head()

air_precipitation.head()
```

```python
air_precipitation['updated_timestamp'] = air_precipitation['date'] + ' ' + air_precipitation['time']

air_precipitation['updated_timestamp'].head()

air_precipitation.to_csv('refined_y.csv')
air_precipitation
```
